# Сложность алгоритма

Сложность алгоритма - O(n), так как здесь мы будем пользоваться алгоритмами линейного поиска.

# Краевые кейсы:


* Пустой список
* Все одинаковые значения в списке

#Код

```python
var removeElement = function(nums, val) {
    let i = 0;
    for (let j = 0; j < nums.length; j++) {
        if (nums[j] !== val) {
            nums[i] = nums[j];
            i++;
        }
    }
    return i;
};
```
